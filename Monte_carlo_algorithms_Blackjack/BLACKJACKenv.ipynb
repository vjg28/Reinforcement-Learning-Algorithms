```python
import numpy as np
import gym
```

```python
#get the environment
env= gym.make('Blackjack-v0')
```

```python
# No. of episodes to run
n_episodes=20

#no of iterations to run
n_iter=100
```

```python
def print_observation(observation):
    
    score, dealer_score, usable_ace = observation
    print('Player Score : {} (Usable ace: {}), Dealer Score: {}' .format(score, usable_ace, dealer_score))
    
```

```python
def strategy(observation):
    score, dealer_score, usable_ace=observation
    #Action1:show the next card
    #Action0 : stop showing cards
    #if score >=20 ,return action=0 else return action=1
    
    return 0 if score >=20 else 1
```

```python
l=1
for i in range(n_episodes):
    observation=env.reset()
    #just to understand how gym env gives output (tuples)
    if l:
        print(observation)
    l=0
    
    print(i)
    
    for number in range(n_iter):
        
        print_observation(observation)
        action=strategy(observation)
        print("Taking action {}".format(["Stick","Hit"][action]))
        observation, reward, done, _ =env.step(action)
        
        if done:
            print_observation(observation)
            print("Game Ended.  Reward = {} \n".format(float(reward)))
            break
```
