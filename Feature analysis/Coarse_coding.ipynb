```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def squareWaveValue(i):
    if i>2 and i<4:
        return 1
    else:
        return 0
```

```python
def feature_definer(feature_numbers, feature_width):
    a=np.random.uniform(0,6-feature_width,feature_numbers)
    b=a+ feature_width
    c=[(a[i],b[i]) for i in range(feature_numbers)]
    return c
```

```python
def feature_vec(feature_ranges,state):
    feature= np.zeros(len(feature_ranges))
    k=0
    for i, j in feature_ranges:
        if state > i and state < j:
            feature[k]=1
        k +=1
    return feature
```

```python
def function_predictor(n_iterations,feature_numbers, feature_width, step_size):
    W = np.random.rand(feature_numbers)*2 -1
    intervals = feature_definer(feature_numbers,feature_width)
    for i in range(n_iterations):
        state = np.random.uniform(0,6)
        #print(state)
        v_sample = squareWaveValue(state)
        #print(v_sample)
        feature= feature_vec(intervals, state)
        #print(feature)
        v_estimate = np.dot(feature, W)
        #print(v_estimate)
        
        #Update of Weights
        W += step_size* (v_sample-v_estimate) * feature
        #print(W)
        #print("\n")
        #step_size *= 1/(i+1)
    
    x_axis = np.arange(-1,7,0.0001)
    function= np.multiply(x_axis<4, x_axis >2)
    plt.plot(x_axis, function)
    plt.show()
    
    i=0
    y_axis=np.zeros(len(x_axis))
    for a in x_axis:
        y_axis[i]= np.dot(W, feature_vec(intervals, a))
        i += 1
    plt.plot(x_axis,y_axis)
    plt.show()
```

```python
function_predictor(n_iterations=1000000, feature_numbers=50, feature_width=2, step_size=0.004)
```
