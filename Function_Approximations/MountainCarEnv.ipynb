```python
import gym
import numpy as np
from matplotlib import pyplot as plt
```

```python
env= gym.make('MountainCar-v0')

a=env.reset()
print(a)
print("\n")
plt.figure()
plt.imshow(env.render(mode='rgb_array'))

for x in range(500):
    
    #There are three actions 0, 1, 2. 0 == accelerate left, 2== accelerate right 1== zero acceleration
    a=env.step(2)
    
    #env.step() returns [(position, velocity), reward, done]
    print(a,x)
    
    #plt.figure()
    plt.imshow(env.render(mode='rgb_array'))

env.close()
```
